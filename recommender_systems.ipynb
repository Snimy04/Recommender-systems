```python
user_book_df.fillna(0,inplace=True)
user_book_df
```

```python
#cosine similarity and correlation between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim=1-pairwise_distances(user_book_df.values,metric='cosine')
user_sim
```

```python
#dataframe of cosine similarities
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
#set index and column names as userid
user_sim_df.index=book1.UserID.unique()
user_sim_df.columns=book1.UserID.unique()
user_sim_df.iloc[0:10,0:10]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:10,0:10]
```

```python
#most similar users
user_sim_df.idxmax(axis=1)[0:20]
```
